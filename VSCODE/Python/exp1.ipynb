```python
import numpy as np
import pandas as pd
import statistics as st
```

```python
 data= np.array([15,45,56,1,2,4,666,64,88,89,78])
```

```python
print(st.mean(data))
print(st.mode(data))
print(st.median(data))
print(np.sort(data))
print(st.stdev(data.tolist()))  # Convert to list
print(st.variance(data.tolist()))    # Convert to list
print(np.max(data))
print(np.min(data))
print(np.sum(data))
print(st.harmonic_mean(data.tolist()))  # Convert to list
print(np.percentile(data, 25))
print(st.median_low(data.tolist()))  # Changed from np.median_low
print(st.median_high(data.tolist())) # Changed from np.median_high
```

```python
from scipy.stats import skew
print(skew(data))
import scipy
scipy.stats.kurtosis(data)
```
