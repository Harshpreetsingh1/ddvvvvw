```python
from PIL import Image, ImageDraw, ImageFont
import textwrap

# Load the base image (image_3.png)
base_image_path = r"C:\VS CODE PROGRAMS\.vscode\Python\WhatsApp Image 2026-01-03 at 11.48.01.jpeg"
try:
    image = Image.open(base_image_path).convert("RGB")
except FileNotFoundError:
    print(f"Error: {base_image_path} not found. Please ensure the file is in the correct path.")
    exit()

draw = ImageDraw.Draw(image)

# Define fonts (using a standard serif font to match the style)
# You may need to adjust the path to your system's fonts
try:
    # Using a generic serif font name, Pillow will search for it.
    # If not found, it will fall back to a default font.
    # For more precise control, specify the full path to a .ttf file.
    # e.g., font_path = "/usr/share/fonts/truetype/msttcorefonts/Times_New_Roman.ttf"
    font_regular = ImageFont.truetype("times.ttf", 24)
    font_bold = ImageFont.truetype("timesbd.ttf", 28)
    font_date = ImageFont.truetype("times.ttf", 22)
except OSError:
    print("Warning: Times New Roman font not found. Falling back to default font.")
    font_regular = ImageFont.load_default()
    font_bold = ImageFont.load_default()
    font_date = ImageFont.load_default()

# Define text content with grammar and spelling corrections
date_text = "03/01/2026"
heading_text = "TO WHOM IT MAY CONCERN"
body_text = (
    "This is to certify that Mr. Sahibjit Singh, son of Amandeep Singh, "
    "has been officially selected to represent Punjab in the 75th Senior National "
    "Basketball Championship, scheduled to be held at Chennai, Tamil Nadu, "
    "from 4th January to 11th January 2026."
)

# Define text positions and margins
date_position = (750, 310)
heading_y_position = 400
body_start_y_position = 450
left_margin = 100
right_margin = 100
image_width = image.width
text_width = image_width - left_margin - right_margin

# Draw the date (centered)
date_bbox = draw.textbbox((0, 0), date_text, font=font_date)
date_width = date_bbox[2] - date_bbox[0]
date_x_position = (image_width - date_width) // 2
draw.text((date_x_position, date_position[1]), date_text, font=font_date, fill="black")

# Draw the heading (centered and underlined)
# Calculate heading width to center it
heading_bbox = draw.textbbox((0, 0), heading_text, font=font_bold)
heading_width = heading_bbox[2] - heading_bbox[0]
heading_x_position = (image_width - heading_width) // 2
draw.text((heading_x_position, heading_y_position), heading_text, font=font_bold, fill="black")

# Draw underline for the heading
underline_y = heading_y_position + heading_bbox[3] + 2 # Add a small padding
draw.line((heading_x_position, underline_y, heading_x_position + heading_width, underline_y), fill="black", width=2)

# Draw the body text (centered, wrapped)
# Use textwrap to fit the text within the margins
wrapper = textwrap.TextWrapper(width=80) # Adjust width as needed for the font size
wrapped_text = wrapper.wrap(body_text)

current_y = body_start_y_position
for line in wrapped_text:
    line_bbox = draw.textbbox((0, 0), line, font=font_regular)
    line_width = line_bbox[2] - line_bbox[0]
    line_x_position = (image_width - line_width) // 2
    draw.text((line_x_position, current_y), line, font=font_regular, fill="black")
    # Calculate line height for spacing
    line_height = line_bbox[3] - line_bbox[1]
    current_y += line_height + 10 # Add some spacing between lines

# Save the final image
output_image_path = "generated_document.png"
image.save(output_image_path)
print(f"Document generated and saved as {output_image_path}")

# Optionally display the image
# image.show()
```
